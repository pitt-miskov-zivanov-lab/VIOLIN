### Embeddings from chatgpt

```python
client = OpenAI()
```

```python
list1 = ["IL2", "rna", "microRNA", "IL-2", "IL6", 'protein family']
list2 = ["Interleukin 2", "protein", "RNA", "Interleukin 6", 'IL-2', 'Protein']
```

```python
for entity1, entity2 in zip(list1, list2):
    e1 = client.embeddings.create(
    model='text-embedding-ada-002',
    input= entity1
)
    e2 = client.embeddings.create(
    model='text-embedding-ada-002',
    input= entity2
)
    similarity = cosine(list(e1.data[0])[0][1], list(e2.data[0])[0][1])
    match = 'True' if similarity <= 0.1 else 'False'
    print(f'{entity1}, {entity2}: {1 - similarity}, {match}')
```

```python
for entity1 in list1:
    for entity2 in list2:
        e1 = client.embeddings.create(
    model='text-embedding-ada-002',
    input= entity1
)
        e2 = client.embeddings.create(
    model='text-embedding-ada-002',
    input= entity2
)
        similarity = cosine(list(e1.data[0])[0][1], list(e2.data[0])[0][1])
        match = 'True' if similarity <= 0.1 else 'False'
        print(f'{entity1}, {entity2}: {1 - similarity}, {match}')
```
